```python
import pandas as pd

baby_names = pd.read_csv("datasets/Popular_Baby_Names.csv")
```

```python
bndf = baby_names[baby_names["Gender"] == "FEMALE"].groupby(["Year of Birth", "Child's First Name"]).sum()

df = bndf.rename_axis(index={"Year of Birth": "YOB", "Child's First Name":"Name"})


df
```

```python
df["My_Rank"] = df.groupby('YOB')['Count'].rank(ascending=False)

df[df["My_Rank"] == 1]
```
